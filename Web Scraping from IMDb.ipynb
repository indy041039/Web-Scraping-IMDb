```python
title = []
year = []
rate = []
time = []
genre = []
score = []
metascore = []
plot = []
director = []
stars = []
votes = []
pic_link = []
gross = []
movie_df = pd.DataFrame()
for movie in movie_div:
    
    # Name
    title_ = movie.find(class_='lister-item-header').a.get_text()
    title.append(title_)
    
    # Release Year
    year_ = movie.find(class_='lister-item-year text-muted unbold').get_text()
    year.append(year_)
    
    # Rate
    rate_ = movie.find(class_='certificate').get_text() if movie.find(class_='certificate') else None
    rate.append(rate_)
    
    # Runtime
    time_ = movie.find(class_='runtime').get_text()
    time.append(time_)
    
    # Genre
    genre_ = movie.find(class_='genre').get_text().strip()
    genre.append(genre_)
    
    # Score
    score_ = movie.find(class_='inline-block ratings-imdb-rating').get_text().strip()
    score.append(score_)
    
    # Metascore
    metascore_ = movie.find(class_='metascore').get_text().strip() if movie.find(class_='metascore') else None
    metascore.append(metascore_)
    
    # Plot 
    plot_ = movie.find_all(class_='text-muted')[2].get_text().strip()
    plot.append(plot_)
    
    # Dicrector 
    director_ = movie.find('p', class_='').get_text().split('|')[0].strip().replace('\n','')[9:]
    director.append(director_)
    
    # Stars
    stars_ = movie.find('p', class_='').get_text().split('|')[1].strip().replace('\n','')[6:]
    stars.append(stars_)
    
    # Vote and Gross
    regex1 = re.compile(r'Votes:(\S*)')
    regex2 = re.compile(r'Gross:(.*)')
    votes_gross = movie.find('p', class_='sort-num_votes-visible').get_text().strip().replace('\n','')    
    votes_ = regex1.search(votes_gross).group(1)[:-1]
    votes.append(votes_)
    gross_ = regex2.search(votes_gross).group(1) if regex2.search(votes_gross) else None
    gross.append(gross_)
```

```python
movie_df = pd.DataFrame({'Name': title,
                        'Release Year': year,
                        'Rate': rate,
                        'Runtime': time,
                        'Genre': genre,
                        'IMDb Score': score,
                        'Metascore': metascore,
                        'Plot': plot,
                        'Director': director,
                        'Stars': stars,
                        'Votes': votes,
                        'Gross': gross})
movie_df
```

```python
from selenium import webdriver
```

```python
browser = webdriver.Chrome()
browser.get('https://www.imdb.com/search/title/?groups=top_1000')

title = []
year = []
rate = []
time = []
genre = []
score = []
metascore = []
plot = []
director = []
stars = []
votes = []
pic_link = []
gross = []
movie_df = pd.DataFrame()

while True:
    r = requests.get(browser.current_url, headers = {'Accept-Language':'en-US, en;q=0.5'})
    soup = BeautifulSoup(r.text, 'html.parser')
    movie_div = soup.find_all('div', {'class' : 'lister-item mode-advanced'})
    
    for movie in movie_div:
    
        # Name
        title_ = movie.find(class_='lister-item-header').a.get_text()
        title.append(title_)
    
        # Release Year
        year_ = movie.find(class_='lister-item-year text-muted unbold').get_text()[1:5]
        year.append(year_)
    
        # Rate
        rate_ = movie.find(class_='certificate').get_text() if movie.find(class_='certificate') else None
        rate.append(rate_)
    
        # Runtime
        time_ = movie.find(class_='runtime').get_text()
        time.append(time_)
    
        # Genre
        genre_ = movie.find(class_='genre').get_text().strip()
        genre.append(genre_)
    
        # Score
        score_ = movie.find(class_='inline-block ratings-imdb-rating').get_text().strip()
        score.append(score_)
    
        # Metascore
        metascore_ = movie.find(class_='metascore').get_text().strip() if movie.find(class_='metascore') else None
        metascore.append(metascore_)
    
        # Plot 
        plot_ = movie.find_all(class_='text-muted')[2].get_text().strip()
        plot.append(plot_)
    
        # Dicrector 
        director_ = movie.find('p', class_='').get_text().split('|')[0].strip().replace('\n','')[9:]
        director.append(director_)
    
        # Stars
        stars_ = movie.find('p', class_='').get_text().split('|')[1].strip().replace('\n','')[6:]
        stars.append(stars_)
    
        # Vote and Gross
        regex1 = re.compile(r'Votes:(\S*)')
        regex2 = re.compile(r'Gross:(.*)')
        votes_gross = movie.find('p', class_='sort-num_votes-visible').get_text().strip().replace('\n','')    
        votes_ = regex1.search(votes_gross).group(1)[:-1]
        votes.append(votes_)
        gross_ = regex2.search(votes_gross).group(1) if regex2.search(votes_gross) else None
        gross.append(gross_)
    timetime.sleep(3)     
    try:
        browser.find_element_by_link_text('Next »').click()
    except:
        break
    timetime.sleep(1)
```

```python
movie_df = pd.DataFrame({'Name': title,
                        'Release Year': year,
                        'Rate': rate,
                        'Runtime': time,
                        'Genre': genre,
                        'IMDb Score': score,
                        'Metascore': metascore,
                        'Plot': plot,
                        'Director': director,
                        'Stars': stars,
                        'Votes': votes,
                        'Gross': gross})
movie_df
```
